```python
import matplotlib.pyplot as plt 
import pandas as pd 

```

```python
df = pd.read_csv("Airbnb_Open_Data.csv", low_memory=False)
df.head(5)
```

```python
# What I like to do is to check what data am I dealing with
df.info()
```

```python
df.isnull().sum() # Also check for null values, so if column have too many null values, I will drop them
df_old = df.copy() # I will create a copy of the dataframe, so I can work with it without changing the original one
```

```python
# check if column have more than 80% null values and if so I drop them (also depeneds on what type of data it is) 
df.dropna(thresh=len(df)*0.8, axis=1, inplace=True) # drop columns with more than 80% null values
```

```python
df.isnull().sum()
```

```python
df_old.columns.difference(df.columns) # check which columns have disappeared
```

```python
df
```

```python
df[['country code','country']].value_counts() 
# We can see that the only country is United States, so we can fill the missing country code with US 
```

```python
df['country code'].fillna('US', inplace=True) # fill missing values with US
df['country'].fillna('United States', inplace=True) # fill missing values with United States
df.isnull().sum() # check if there are still null values
```

```python
df['country code'].isna().sum() # done
```

```python
# Also the Contruction Year column is float, I assume it should be date format. Lets fix that
print(df['Construction year'].dtype)

df['Construction year'] = df['Construction year'].astype('Int64')  
```

```python
df
```

```python
#Latitude (lat) typically corresponds to the y-coordinate, representing the north-south position.
#Longitude (long) typically corresponds to the x-coordinate, representing the east-west position.
# I want geographic coordinate for later usage on powebi visualization 
df['geo_coordinate'] = list(zip(df['long'], df['lat']))
```

```python
df['geo_coordinate'] 
```

```python
# Price is a string so i will convert it to float, but first i will remove the $ sign 
df['price'] = df['price'].str.replace('$', '').str.replace(',', '').astype('float64')
df['price'].head(2)
# The same with service fee
df['service fee'] = df['service fee'].str.replace('$', '').str.replace(',', '').astype('float64')
df['service fee'].head(2)
```

```python
# The rest we can do in PowerBI, so I will save the dataframe to csv file
df.to_csv('Airbnb_Open_Data_Processed.csv', index=False, sep=';')
```
